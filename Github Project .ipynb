### Indian Tourism Statewise Statistics [2019 - 2020]

#### Foreign tourists per State/UTs [2019 - 2020]

```python
state_foregin_tourist = pd.read_csv("Downloads/India-Tourism-Data-Analysis-Project-py-main/India-Tourism-Data-Analysis-Project-py-main/Data set/India-Tourism-Statistics-statewise_2019-2020_domestic_foreign.csv")
state_foregin_tourist
```

```python
foreign_Tourist=state_foregin_tourist[['States/UTs','Foreign - 2019','Foreign - 2020']]
foreign_Tourist
```

```python
# sorting 
foreign_Tourist.sort_values(by=['Foreign - 2019'], ascending=False)
```

```python
foreign_Tourist['Foreign - 2019'] = foreign_Tourist['Foreign - 2019']*10**-4
```

```python
foreign_Tourist[foreign_Tourist['States/UTs'].isin(['Grand Total'])==False]
```

```python
foreign_Tourist.sort_values(by=['Foreign - 2019'], ascending=False)
df1=pd.melt(foreign_Tourist,['States/UTs']).rename(columns={"variable":"Year"})
plt.figure(figsize=(14,40))
ax = sns.barplot(x="value", y="States/UTs", data=df1, hue='Year')
for p in ax.patches:
    width = p.get_width()    
    ax.text(width + 0.5,       
            p.get_y() + p.get_height() / 2, 
            '{:1.2f}'.format(width), 
            ha = 'left',   
            va = 'center',  
            fontsize=10)
        
```

#### Domestic tourists per State/UTs [2019 - 2020]

```python
domestic=state_foregin_tourist[['States/UTs','Domestic -2019','Domestic -2020']].copy()
# sorting by Domestic-2019 in descending order
domestic.sort_values(by=['Domestic -2019'], ascending=False, inplace=True)
domestic=domestic[domestic['States/UTs'].isin(['Grand Total'])==False]
domestic
```

```python
domestic['Domestic -2019'] = domestic['Domestic -2019']*10**-4
```

```python
domestic['Domestic -2020'] = domestic['Domestic -2020']*10**-4
domestic
```

```python
df2=pd.melt(domestic,['States/UTs']).rename(columns={"variable":"Year"})
plt.figure(figsize=(14,40))
ax = sns.barplot(x="value", y="States/UTs", data=df2, hue='Year')
for p in ax.patches:
    width = p.get_width()    
    ax.text(width + 0.5,       
            p.get_y() + p.get_height() / 2, 
            '{:1.2f}'.format(width), 
            ha = 'left',   
            va = 'center',  
            fontsize=10)
        
```

## Indian Popular monuments during 2019,2020,2021
